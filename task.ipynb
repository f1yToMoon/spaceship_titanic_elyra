```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import category_encoders as ce
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
train_df = pd.read_csv("C:/Users/Никита/DS-task/train.csv")
train_df.head()
```

### Удалим столбцы PassengerId и Name

```python
train_df.drop(
    ['PassengerId', 'Name'],
    axis=1,
    inplace=True
)
train_df.head()
```

```python
train_df.info()
```

### Поменяем тип столбцов CryoSleep и VIP на bool:

```python
columns_to_bool = ["CryoSleep", "VIP"]
train_df[columns_to_bool] = train_df[columns_to_bool].astype('bool')
```

### В данном случае NaN-ы конвертируются в False, что не является проблемой, так как для этих столбцов False является модой(из описания датасета).

```python
train_df.info()
```

### в столбце Cabin оставим только уровень палубы, те первый символ

```python
for i in range(0 ,len(train_df.Cabin)):
    if not(pd.isna(train_df.Cabin[i])):
        s = str(train_df.Cabin[i])
        train_df.Cabin[i] = s[0]
```

```python
cat_df = train_df.select_dtypes(include=['object'])
```

```python
num_df = train_df.select_dtypes(exclude=['object'])
```

```python
def corrplot(df, method="pearson", annot=True, **kwargs):
    sns.clustermap(
        df.corr(method),
        vmin=-1.0,
        vmax=1.0,
        cmap="icefire",
        method="complete",
        annot=annot,
        **kwargs,
    )


corrplot(num_df, annot=None)
```

### Из этой матрицы можно увидеть, что столбцы CryoSleep и Transported коррелируют между собой, поэтому удаляем столбец CryoSleep

```python
train_df.drop(
    ['CryoSleep'],
    axis=1,
    inplace=True
)
```

### Соберем статистику, связанную с NaN-ми.

```python
nan_df = (train_df.isnull().mean() * 100).reset_index()
nan_df.columns = ["column_name", "percentage"]
nan_df.sort_values("percentage", ascending=False, inplace=True)
nan_df
```

```python
df = train_df.dropna()
print("если удалять строки с NaN-ми будет потеряно следующее количество строк: ", train_df.shape[0] - df.shape[0])
```

###  Удалять такое количество строк не представляется возможным, поэтому заполним NaN-ы для стобца Age средним значением, а для столбцов RoomService, FoodCourt, ShoppingMall, Spa и VRDeck NaN-ы заполним модой(такое решение было принято основываясь на гистограммах в описании датасета) 


```python
num_сols_with_missing = [col for col in num_df.columns 
                         if num_df[col].isnull().any()]
num_сols_with_missing
```

```python
train_df['Age'] = train_df['Age'].transform(lambda x: x.fillna(x.mean()))
```

```python
num_сols_with_missing = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
train_df[num_сols_with_missing] = train_df[num_сols_with_missing].transform(lambda x: x.fillna(x.mode()))
```

```python
cat_сols_with_missing = [col for col in cat_df.columns 
                         if cat_df[col].isnull().any()]
cat_сols_with_missing
```

### Столбцы HomePlanet и Destination заполним модой(используем описание датасета).


```python
cat_сols_with_missing = ['HomePlanet', 'Destination']
train_df[cat_сols_with_missing] = train_df[cat_сols_with_missing].transform(lambda x: x.fillna(x.dropna().mode()[0]))
```

### Для столбца Cabin просто удалим строки с NaN-ми(их всего 2%)

```python
train_df = train_df.dropna()
```

### Построим ящики с усами и удалим экстременальные значения, если такие есть.

```python
out_col = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

fig, axes = plt.subplots(1, 6, figsize=(20, 15))
for ax, col in zip(axes, out_col):
    sns.boxplot(y=train_df[col], data=train_df, ax=ax)
fig.tight_layout(pad=1.5)
```

```python
for col, upper_bound in (
    ('RoomService', 5000),
    ('FoodCourt', 20000),
    ('ShoppingMall', 5000),
    ('Spa', 10000),
    ('VRDeck', 14000),
):
    train_df = train_df.drop(train_df[train_df[col] > upper_bound].index)
```
